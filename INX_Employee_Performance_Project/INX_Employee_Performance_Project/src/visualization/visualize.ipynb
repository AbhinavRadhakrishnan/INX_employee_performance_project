# Feature Importance Analysis

**Objective:**  
Identify top factors influencing employee performance.


```python
import pandas as pd
import matplotlib.pyplot as plt
import joblib
```

```python
df = pd.read_csv("../../data/processed/employee_performance_processed.csv")
model = joblib.load("../models/employee_performance_model.pkl")
```

```python
importance = pd.Series(
    model.feature_importances_,
    index=df.drop("PerformanceRating", axis=1).columns
).sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,5))
importance.head(10).plot(kind="bar")
plt.title("Top Factors Affecting Employee Performance")
plt.ylabel("Importance")
plt.show()
```
