# Exploratory Data Analysis (EDA)

**Objective:**  
Analyze employee performance distribution and department-wise trends.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../data/processed/employee_performance_processed.csv")
```

```python
# Performance distribution
plt.figure(figsize=(6,4))
sns.countplot(x="PerformanceRating", data=df)
plt.title("Performance Rating Distribution")
plt.show()
```

```python
# Department-wise performance (CORRECT COLUMN)
dept_perf = df.groupby("EmpDepartment")["PerformanceRating"].mean().sort_values(ascending=False)
```

```python

plt.figure(figsize=(10,5))
dept_perf.plot(kind="bar")
plt.title("Department-wise Average Performance Rating")
plt.ylabel("Average Rating")
plt.show()
```

```python

# Correlation heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
