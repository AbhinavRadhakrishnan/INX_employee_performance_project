# Employee Performance Prediction

**Objective:**  
Predict performance for a sample employee.


```python
import pandas as pd
import joblib
```

```python
model = joblib.load("employee_performance_model.pkl")
scaler = joblib.load("scaler.pkl")
```

```python
df = pd.read_csv("../../data/processed/employee_performance_processed.csv")
```

```python
sample = df.drop("PerformanceRating", axis=1).iloc[0:1]
sample_scaled = scaler.transform(sample)
```

```python

prediction = model.predict(sample_scaled)
print("Predicted Performance Rating:", prediction[0])
```
